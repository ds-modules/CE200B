## Taylor-Green vortex

```python
# Set the parameters for the Taylor-Green vortex example
Nx = 32 # 64 N = number of cells in x - note: 64 seems to run pretty slowly
Ny = 32  # 64 N = number of cells in y - note: 64 seems to run pretty slowly
Lx = 2*np.pi # L = physical dimension in x
Ly = 2*np.pi # L = physical dimension in y
# dt = min((Lx/Nx),(Ly/Ny))/3000; # dt = timestep
dt = min((Lx/Nx),(Ly/Ny)) / 10; # dt = timestep -- use for Re = 100,1000
T = 1; # T = final time
# nu = 10 for Re 0.1, nu = 0.01 for Re 100, nu = 0.001 for Re 1000
nu = 0.1;
ng = 1; # ng = number of ghostcells

# BC = boundary conditions (4x1 array)
BC = np.array([[complex(0,1),complex(0,1),complex(0,1),complex(0,1)]])

# IC_choice = choice of initial conditions defined in IC.m: Taylor-Green vortex
IC_choice = 3;

# misc
PGx = 0 # pressure gradient in x
PGy = 0 # pressure gradient in y
IBM = 0 # IBM 0 = off, 1 = on
```

```python
# Call the BINS code
BINS(Nx,Ny,Lx,Ly,dt,T,BC,IC_choice,nu,ng,PGx,PGy,IBM)
```
